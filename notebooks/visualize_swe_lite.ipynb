# SWE-bench Lite Dataset Visualization
This notebook loads the SWE-bench Lite dataset from Hugging Face for exploration.

```python
import pandas as pd

# Note: You may need to login to Hugging Face if the dataset requires it:
# !huggingface-cli login

splits = {'dev': 'data/dev-00000-of-00001.parquet', 'test': 'data/test-00000-of-00001.parquet'}
df = pd.read_parquet("hf://datasets/princeton-nlp/SWE-bench_Lite/" + splits["dev"])
df.head()
```

### Column Definitions
- **`instance_id`**: A unique identifier for the issue (e.g., `sqlfluff__sqlfluff-1625`).
- **`repo`**: The repository the issue belongs to (e.g., `sqlfluff/sqlfluff`).
- **`base_commit`**: The commit hash of the repository *before* the issue was fixed. The agent starts here.
- **`problem_statement`**: The text of the bug report or feature request from GitHub.
- **`patch`**: The gold/reference diff that fixes the issue (used for evaluation; hidden from the agent).
- **`test_patch`**: The diff containing the test cases that verify the issue. They fail on `base_commit` but pass on the fixed commit.
- **`version`**: The major/minor version of the repo when the issue was opened.
- **`FAIL_TO_PASS`**: List of test names that fail on the `base_commit` and pass on the reference fix.
- **`PASS_TO_PASS`**: List of test names that pass on both (used to check for regressions).
- **`environment_setup_commit`**: Commit used to build the execution environment.
- **`hints_text`**: Optional hints (from the issue thread). Usually excluded from the agent`s prompt in Lite.
- **`created_at`**: Date the issue was created.
